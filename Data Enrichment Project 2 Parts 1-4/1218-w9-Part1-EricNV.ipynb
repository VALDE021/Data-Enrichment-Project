### 3. Remove non-US movies from title basicst

- Any rows from title basics with tconst id's that are not incuded in the AKA's titled's. You can use panda's .isin() method to remove non-US movies from title basics (example included below).

### 4. Perform the remaining preprocessing steps for title basics.

```python
akas = akas.replace({'\\N' : np.nan})
```

- Filter to keep only full-length movies (titleType==Movie)

```python
filter_type = basics['titleType'] =='movie'
basics = basics[filter_type]
basics
```

- Convert startYear to a float dtype

```python
basics['startYear'] = basics['startYear'].astype(float)
basics.info()
```

- Filter to keep movies with startYears that are >=2000 and <=2022

```python
basics = basics[(basics['startYear'] >= 2000) & (basics['startYear'] <= 2022)]
basics.head()
```

- Eliminate movies that include "Documentary" in genre:
    - You will need to create a filter if the movie's value in the genres column contains the word Documentary.

```python
basics.info()
```

```python
basics.info('genres')
```

```python
#Eliminate movies that are null for runtimeMinute and genres
basics = basics.dropna(subset=['runtimeMinutes','genres'])
basics.isna().sum()
```

```python
filter_documentaries = basics['genres'].str.contains('Documentary',case = False)
filter_documentaries
```

```python
# Exclude movies in the documentary category.
basics = basics[~filter_documentaries]
```

You will also want to use the ~ operator to take the inverse of your Trues/Falses.
